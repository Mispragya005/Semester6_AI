```python
def convert_units(conversion_type, value):
    """
    Converts a value between different units of measurement.

    Parameters:
    conversion_type (str): The type of conversion to perform.
                           Options:
                           - 'm_to_ft'
                           - 'ft_to_m'
                           - 'kg_to_lbs'
                           - 'lbs_to_kg'
                           - 'l_to_gal'
                           - 'gal_to_l'
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    # Length conversions
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084

    # Weight conversions
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462

    # Volume conversions
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172

    else:
        raise ValueError("Unsupported conversion type.")


def main():
    print("Unit Conversion Program")
    print("-----------------------")
    print("Choose a conversion category:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    category = input("Enter your choice (1/2/3): ")

    try:
        if category == "1":
            print("\nLength Conversions:")
            print("a. Meters to Feet")
            print("b. Feet to Meters")
            choice = input("Choose conversion (a/b): ")

            value = float(input("Enter the value to convert: "))

            if choice == "a":
                result = convert_units("m_to_ft", value)
                print(f"{value} meters = {result:.4f} feet")
            elif choice == "b":
                result = convert_units("ft_to_m", value)
                print(f"{value} feet = {result:.4f} meters")
            else:
                print("Invalid conversion choice.")

        elif category == "2":
            print("\nWeight Conversions:")
            print("a. Kilograms to Pounds")
            print("b. Pounds to Kilograms")
            choice = input("Choose conversion (a/b): ")

            value = float(input("Enter the value to convert: "))

            if choice == "a":
                result = convert_units("kg_to_lbs", value)
                print(f"{value} kg = {result:.4f} lbs")
            elif choice == "b":
                result = convert_units("lbs_to_kg", value)
                print(f"{value} lbs = {result:.4f} kg")
            else:
                print("Invalid conversion choice.")

        elif category == "3":
            print("\nVolume Conversions:")
            print("a. Liters to Gallons")
            print("b. Gallons to Liters")
            choice = input("Choose conversion (a/b): ")

            value = float(input("Enter the value to convert: "))

            if choice == "a":
                result = convert_units("l_to_gal", value)
                print(f"{value} L = {result:.4f} gallons")
            elif choice == "b":
                result = convert_units("gal_to_l", value)
                print(f"{value} gallons = {result:.4f} L")
            else:
                print("Invalid conversion choice.")

        else:
            print("Invalid category selection.")

    except ValueError:
        print("Error: Please enter a valid numeric value for conversion.")
    except Exception as e:
        print(f"Unexpected error: {e}")


if __name__ == "__main__":
    main()
```

```python
def perform_operation(numbers, operation):
    """
    Performs a mathematical operation on a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers to process.
    operation (str): The operation to perform.
                     Options:
                     - 'sum'
                     - 'average'
                     - 'max'
                     - 'min'

    Returns:
    float: The result of the selected operation.

    Raises:
    ValueError: If the list is empty or the operation is unsupported.
    """

    if not numbers:
        raise ValueError("The number list is empty.")

    if operation == "sum":
        return sum(numbers)
    elif operation == "average":
        return sum(numbers) / len(numbers)
    elif operation == "max":
        return max(numbers)
    elif operation == "min":
        return min(numbers)
    else:
        raise ValueError("Unsupported operation.")


def main():
    print("Mathematical Operations Program")
    print("--------------------------------")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1/2/3/4): ")

    try:
        user_input = input("Enter numbers separated by spaces: ")

        # Convert input string into a list of floats
        numbers = [float(num) for num in user_input.split()]

        if choice == "1":
            result = perform_operation(numbers, "sum")
            print(f"Sum = {result}")
        elif choice == "2":
            result = perform_operation(numbers, "average")
            print(f"Average = {result}")
        elif choice == "3":
            result = perform_operation(numbers, "max")
            print(f"Maximum = {result}")
        elif choice == "4":
            result = perform_operation(numbers, "min")
            print(f"Minimum = {result}")
        else:
            print("Invalid operation choice.")

    except ValueError as ve:
        print(f"Error: {ve}")
    except Exception:
        print("Invalid input. Please enter numeric values only.")


if __name__ == "__main__":
    main()
```

```python
def extract_every_other(lst):
    """
    Returns a new list containing every other element
    from the input list, starting with the first element.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list containing every other element.
    """
    return lst[::2]


# Example usage
example = [1, 2, 3, 4, 5, 6]
result = extract_every_other(example)
print(result)  # Output: [1, 3, 5]
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list, starting at index 'start'
    and ending at index 'end' (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): The starting index.
    end (int): The ending index (inclusive).

    Returns:
    list: The extracted sublist.

    Raises:
    ValueError: If indices are invalid.
    """
    if start < 0 or end >= len(lst) or start > end:
        raise ValueError("Invalid start or end index.")

    return lst[start:end + 1]


# Example usage
example = [10, 20, 30, 40, 50, 60]
result = get_sublist(example, 1, 4)
print(result)  # Output: [20, 30, 40, 50]
```

```python
def reverse_list(lst):
    """
    Returns a new list that is the reverse of the input list
    using slicing.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list with elements in reverse order.
    """
    return lst[::-1]


# Example usage
example = [1, 2, 3, 4, 5]
result = reverse_list(example)
print(result)  # Output: [5, 4, 3, 2, 1]
```

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)  # inverse of A
identity = np.dot(A, A_inv)
print("\n1. A * A^-1 =\n", identity)

# Check if identity is approximately equal to np.eye(2)
print("Is identity approximately equal to I?", np.allclose(identity, np.eye(2)))

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB =\n", AB)
print("   BA =\n", BA)
print("AB == BA?", np.array_equal(AB, BA))  # should be False

# 3. Prove (AB)^T = B^T A^T
AB_T = (AB).T
B_T_A_T = np.dot(B.T, A.T)
print("\n3. (AB)^T =\n", AB_T)
print("   B^T A^T =\n", B_T_A_T)
print("Is (AB)^T equal to B^T A^T?", np.array_equal(AB_T, B_T_A_T))

# ---------------------------
# Solve system of linear equations
# Equations:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

# Coefficient matrix
coeff_matrix = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

# Right-hand side vector
rhs = np.array([-1, -3, 9])

# Solve using inverse method: X = A^-1 * B
coeff_inv = np.linalg.inv(coeff_matrix)
solution = np.dot(coeff_inv, rhs)
print("\nSolution using inverse method:", solution)

# Alternatively, NumPy provides a direct solver
solution_direct = np.linalg.solve(coeff_matrix, rhs)
print("Solution using np.linalg.solve:", solution_direct)
```
